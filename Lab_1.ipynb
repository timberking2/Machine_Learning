##Типизация
Присвойте соответствующим колонкам один из типов (category, float, int)

```python
for col in df.columns:
  if col == 'City':
    continue
  df[col] = df[col].astype(str).str.replace(',','')
for col in df.columns:
  if col == 'City':
    df[col].astype("category")
    continue
  df[col] = df[col].astype('Float64')

df.dtypes
```
